```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
columns = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheeels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
cars = pd.read_csv('imports-85.data',names = columns)
```

```python
cars.head()
```

```python
numeric_columns = ['normalized-losses','wheel-base','length','width','height','curb-weight','engine-size','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']

numeric_cars = cars[numeric_columns]
```

```python
numeric_cars.head()
```

```python
numeric_cars = numeric_cars.replace('?', np.nan)
```

```python
numeric_cars = numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

```python
## predicting "price" column so need to get rid of the empty rows in the price column
numeric_cars = numeric_cars.dropna(subset = ['price'])
numeric_cars.isnull().sum()
```

```python
## replace other missing values with mean
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
```

```python
## normalize all columns except "price"
price_column = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max()-numeric_cars.min())
numeric_cars['price'] = price_column
```

```python
numeric_cars.head()
```

```python
def knn_train_test(training_column, target_column, df):
    knn = KNeighborsRegressor()
    np.random.seed(1)
    
    ## randomize data frame
    shuffled_index = np.random.permutation(df.index)
    df = df.reindex(shuffled_index)
    
    ## splitting the data frame to a training and test set
    last_row = int(len(df)/2)
    train_df = df.iloc[0:last_row]
    test_df = df.iloc[last_row:]
    
    ## fitting the model
    knn.fit(train_df[[training_column]], train_df[target_column])
    
    ## predictions on the test set
    predictions = knn.predict(test_df[[training_column]])
    
    ## calculate the rmse
    mse = mean_squared_error(predictions, test_df[target_column])
    rmse = mse**(1/2)
    return rmse

results = {}
train_columns = numeric_cars.columns.drop('price')

for i in train_columns:
    rmse_value = knn_train_test(i, 'price', numeric_cars)
    results[i] = rmse_value

rmse_series = pd.Series(results)
rmse_series.sort_values()
```

```python
## writing a new function to accept different k values
def knn_train_test(training_column, target_column, df):
    np.random.seed(1)
    
    ## randomize data frame
    shuffled_index = np.random.permutation(df.index)
    df = df.reindex(shuffled_index)
    
    ## splitting the data frame to a training and test set
    last_row = int(len(df)/2)
    train_df = df.iloc[0:last_row]
    test_df = df.iloc[last_row:]
    
    ## training the model and making predictions
    k_values = [1,3,5,7,9]
    rmse_values = {}
    
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit(train_df[[training_column]], train_df[target_column])
        
        predictions = knn.predict(test_df[[training_column]])
        
        mse = mean_squared_error(predictions, test_df[target_column])
        rmse = np.sqrt(mse)
        
        rmse_values[k] = rmse
    return rmse_values            
  
multiple_k_value_rmse = {}
training_columns = numeric_cars.columns.drop('price')

for i in training_columns:
    rmse_value = knn_train_test(i, 'price', numeric_cars)
    multiple_k_value_rmse[i] = rmse_value

print(multiple_k_value_rmse)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k,v in multiple_k_value_rmse.items():
    x = list(v.keys())
    y = list(v.values())
    
    plt.plot(x,y)
    plt.xlabel('K value')
    plt.ylabel('RMSE')
    
## We can see that the very best results are at a k value of 3
```

```python
## determining the exact rmse value on average for each feature
feature_avg_rmse = {}
for k,v in multiple_k_value_rmse.items():
    avg_rmse = np.mean(list(v.values()))
    feature_avg_rmse[k] = avg_rmse
series_avg_rmse = pd.Series(feature_avg_rmse)
sorted_series_avg_rmse = series_avg_rmse.sort_values()
print(sorted_series_avg_rmse)

sorted_features = sorted_series_avg_rmse.index
```
