```python
import pandas as pd
loans_2007 = pd.read_csv('loans_2007.csv')
loans_2007.head()
```

```python
## getting rid of columns that
## leak information from the future (after the loan has already been funded)
## don't affect a borrower's ability to pay back a loan (e.g. a randomly generated ID value by Lending Club)
## contain redundant information
loans_2007 = loans_2007.drop(columns = ['id','member_id','funded_amnt','funded_amnt_inv','grade','sub_grade','emp_title','issue_d'])
```

```python
## getting rid of columns that
## leak information from the future (after the loan has already been funded)
## don't affect a borrower's ability to pay back a loan (e.g. a randomly generated ID value by Lending Club)
## contain redundant information
loans_2007 = loans_2007.drop(columns = ['zip_code','out_prncp','out_prncp_inv','total_pymnt','total_pymnt_inv','total_rec_prncp'])
```

```python
## getting rid of columns that
## leak information from the future (after the loan has already been funded)
## don't affect a borrower's ability to pay back a loan (e.g. a randomly generated ID value by Lending Club)
## contain redundant information
loans_2007 = loans_2007.drop(columns = ['total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt'])
```

```python
## went from 52 columns to begin with down to 32
loans_2007.head()
```

```python
## the loan_status columns describes if a loan was paid off on time, had delayed payments, or was defaulted on the borrower
## we will use the loan_status column as the target column for the model
## the loan_status column contains text values at the moment
loan_status_values = loans_2007['loan_status'].value_counts()
print(loan_status_values)
```

```python
## only loans that are "fully paid" or "charged off" are fully completed loans
## fully paid means the loan is complete
## charged off means the loan has essentially 0 chance of being paid
## we will replace all "fully paid" loans with a value of 1 and all "charged off" loans with a value of 0

loans_2007 = loans_2007[(loans_2007['loan_status'] == "Fully Paid") | (loans_2007['loan_status'] == "Charged Off")]

replace = {
    "loan_status":{
        "Fully Paid":1,
        "Charged Off":0,
    }
}
loans_2007 = loans_2007.replace(replace)
```

```python
loans_2007['loan_status'].head()
```

```python
## we will get rid of the columns that only contain 1 unique value so we have less columns to deal with
## we go through each columns and get rid of any "nan" values 
## then if the length of the unique values in that column is 1 we append that to the list of drop_columns and then drop those columns
columns = loans_2007.columns
drop_columns = []
for col in columns:
    col_series = loans_2007[col].dropna().unique()
    if len(col_series) == 1:
        drop_columns.append(col)
loans_2007 = loans_2007.drop(drop_columns, axis=1)
print(drop_columns)
```
