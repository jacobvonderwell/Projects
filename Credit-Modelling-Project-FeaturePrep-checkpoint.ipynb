```python
import pandas as pd
loans = pd.read_csv('cleaned_credit_data')
loans = loans.drop(columns = ['Unnamed: 0'])
loans.head()
```

```python
## determining the number of null values there are in the dataframe
null_counts = loans.isnull().sum()
print(null_counts[null_counts>0])
```

```python
## the two main columns with a lot of missing values is "emp_length" and "pub_rec_bankruptcies"
## we can assume emp_length will be a good indicator of if someone can pay back a loan or not so we will keep that column for sure
print(loans.pub_rec_bankruptcies.value_counts(normalize=True, dropna=False))
```

```python
## since the majority of the values in "pub_rec_bankruptcies" are the same we can just get rid of that column
loans = loans.drop(columns = ['pub_rec_bankruptcies'])
```

```python
## for the other missing values we will just get rid of the rows that have any missing values
loans = loans.dropna(axis = 0)
print(loans.dtypes.value_counts())
```

```python
## the object types need to be converted to numerical values for ski-kit-learn
## first we will see what the different object values look like
object_columns_df = loans.select_dtypes(include=["object"])
object_columns_df.head()
```

```python
## only some of the columns contain categorical information
cols = ['home_ownership', 'verification_status', 'emp_length', 'term', 'addr_state']
for c in cols:
    print(loans[c].value_counts())
```

```python
## now looking at the "title" and "purpose" columns and their values
print(loans["title"].value_counts())
print(loans["purpose"].value_counts())
```

```python
## we will change the 'emp_length' column with the numerical values listed in the mapping_dict
## we also are going to get rid of "last_credit_pull_d", "earliest_cr_line", "addr_state", "title" since their information is not very useful
mapping_dict = {
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
        "n/a": 0
    }
}
loans = loans.drop(["last_credit_pull_d", "earliest_cr_line", "addr_state", "title"], axis=1)
loans["int_rate"] = loans["int_rate"].str.rstrip("%").astype("float")
loans["revol_util"] = loans["revol_util"].str.rstrip("%").astype("float")
loans = loans.replace(mapping_dict)
```

```python
## we now use dummy variables for the categorical columns so we can get numerical values for the model
cat_columns = ["home_ownership", "verification_status", "purpose", "term"]
dummy_df = pd.get_dummies(loans[cat_columns])
loans = pd.concat([loans, dummy_df], axis=1)
loans = loans.drop(cat_columns, axis=1)
```

```python
loans.to_csv('cleaned_loans_2007')
```
