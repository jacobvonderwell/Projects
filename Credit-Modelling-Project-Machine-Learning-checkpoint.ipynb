```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict
import pandas as pd
loans = pd.read_csv('cleaned_loans_2007')
loans = loans.drop(columns = ['Unnamed: 0'])
loans.head()
```

```python
## creating a machine learning model to make predictions
## We use "loan_status" as the target column and every other column as the training columns
## we set the class_weight to balanced because there is almost a 6:1 ratio of the loan being payed back on time vs not being payed back on time in the data
penalty = {
    0: 10,
    1: 1
}
lr = LogisticRegression(class_weight=penalty)
cols = loans.columns
train_cols = cols.drop("loan_status")
features = loans[train_cols]
target = loans["loan_status"]
lr.fit(features, target)
```

```python
## we now use cross validation to determine the accuracy of the model
predictions = cross_val_predict(lr, features, target, cv=3)
predictions = pd.Series(predictions)
```

```python

## False positives.
fp_filter = (predictions == 1) & (loans["loan_status"] == 0)
fp = len(predictions[fp_filter])

## True positives.
tp_filter = (predictions == 1) & (loans["loan_status"] == 1)
tp = len(predictions[tp_filter])

## False negatives.
fn_filter = (predictions == 0) & (loans["loan_status"] == 1)
fn = len(predictions[fn_filter])

## True negatives
tn_filter = (predictions == 0) & (loans["loan_status"] == 0)
tn = len(predictions[tn_filter])
## Rates
## False positive rate is the number of false positives divided by the number of false positives plus the number of true negatives
##This divides all the cases where we thought a loan would be paid off but it wasn't by all the loans that weren't paid off
## True positive rate is the number of true positives divided by the number of true positives plus the number of false negatives
## This divides all the cases where we thought a loan would be paid off and it was by all the loans that were paid off
tpr = tp  / (tp + fn)
fpr = fp  / (fp + tn)
print(tpr)
print(fpr)
```
